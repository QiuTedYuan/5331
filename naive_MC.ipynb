```python
import random
import math

default_data = "./data/test_gnm/"
with open(default_data+'attribute.txt') as f:
    attr=f.read().split('\n')
n = int(attr[0].split('=')[1])
m = int(attr[1].split('=')[1])
s = 1
alpha = 0.2
pf = 1./n
delta = 1./n
epsilon = 0.5

w = int(math.ceil(math.log(1./pf)/(epsilon*epsilon*delta)))
    
with open(default_data+'graph.txt') as f:
    _graph=f.readlines() 
graph=[[] for i in range(n+1)]
for line in _graph:
    edge = map(int, line.split())
    graph[edge[0]].append(edge[1])

frequency = [0 for i in range(n+1)]
#random_walk
for i in range(w):
    current=s
    while (random.random()>alpha):
        Nout=len(graph[current])
        if (Nout==0):
            break
        choice = int(math.floor(random.random()*Nout))
        current = graph[current][choice]
    frequency[current]+=1

print frequency

        
        


```
